```python
import cv2
from skimage import feature
import numpy as np
import matplotlib.pyplot as plt
import imutils
from imutils import paths
```

```python
#image1 = cv2.imread('D:/Ariadh/cours/cbir/idb/stex-512-splitted/jpg/Marble/Marble.0012.04.jpg')
image1 = cv2.imread('C:/ariadh/python/idb/Objects/Faces/image_0326.jpg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(image1)
gray1 = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
```

```python
numPoints=8
radius=2
lbp = feature.local_binary_pattern(gray1, numPoints,radius, method="uniform") #nri_uniform
n_bins = int(lbp.max() + 1)
hist, _ = np.histogram(lbp, density=True, bins=n_bins, range=(0, n_bins))
#(hist, _) = np.histogram(lbp.ravel(),bins=np.arange(0, numPoints + 3),range=(0, numPoints + 2))
#hist = hist.astype("float")
#hist /= (hist.sum() + 1e-7)
```
